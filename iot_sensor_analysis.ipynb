```python
# ðŸ“Š IoT Sensor Data Analysis
# Author: Shreyans Magdum
# Description: Simple analysis of temperature and humidity sensor readings

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create sample IoT sensor data directly (no need for CSV file)
data = {
    "Time": ["10:00","10:10","10:20","10:30","10:40","10:50","11:00"],
    "Temperature": [22,23,25,24,26,27,28],
    "Humidity": [60,58,55,57,54,52,50]
}

df = pd.DataFrame(data)
```

```python
# Display first few records
print("Sample Sensor Data:")
print(df.head())
```

```python
# Calculate summary statistics
print("\nAverage Temperature:", df["Temperature"].mean())
print("Average Humidity:", df["Humidity"].mean())
```

```python
# Visualize the readings
plt.figure(figsize=(8,5))
plt.plot(df["Time"], df["Temperature"], marker='o', label="Temperature (Â°C)")
plt.plot(df["Time"], df["Humidity"], marker='x', label="Humidity (%)")
plt.title("IoT Sensor Data Monitoring")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()
```
